#  Framework

<h2><center>Arquitetura: </center></h2>
<center>
<img src= imgs/overview.png
<center>


- __Runtime Dependency Engine__: Schedules and executes the operations according to their read/write dependency.
- __Storage Allocator__: Efficiently allocates and recycles memory blocks on host (CPU) and devices (GPUs).
- __Resource Manager__: Manages global resources, such as the random number generator and temporal space.
- __NDArray__: Dynamic, asynchronous n-dimensional arrays, which provide flexible imperative programs for MXNet.
- __Symbolic Execution__: Static symbolic graph executor, which provides efficient symbolic graph execution and optimization.
- __Operator__: Operators that define static forward and gradient calculation (backprop).
- __SimpleOp__: Operators that extend NDArray operators and symbolic operators in a unified fashion.
- __Symbol Construction__: Symbolic construction, which provides a way to construct a computation graph (net configuration).
- __KVStore__: Key-value store interface for efficient parameter synchronization.
- __Data Loading(IO)__: Efficient distributed data loading and augmentation.


- A partir desses módulos *Core* (cpp) temos 3 apis de alto-nível em Python:
  - __Module__:
   - Com base no módulo Symbol: grafos estáticos (*define, compile, run*);
   - Mais restrito, mas com maior performance;
 - __Keras__;

- __Gluon__:
 - Interface mais nova;
 - Suporte tanto à programação imperativa ( dinâmica, *define by run*) quanto simbólica (__mx.net.hybridize()__);
 - Possui extensões para CV e NLP (gluon-cv, gluon-nlp);

# Imports 

```python
import numpy as np

import keras
from keras.models import Sequential
from keras.layers import Dense

import mxnet as mx
from mxnet import autograd
from mxnet import gluon
import mxnet.ndarray as nd
import mxnet.gluon.nn as mx_nn
import mxnet.gluon.data as gdt
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import sklearn.preprocessing as preprocessing
from sklearn.model_selection import train_test_split
import sklearn


def plot_pie(y):
    from collections import Counter
    target_stats = Counter(y)
    labels = list(target_stats.keys())
    sizes = list(target_stats.values())
    explode = tuple([0.1] * len(target_stats))

    def make_autopct(values):
        def my_autopct(pct):
            total = sum(values)
            val = int(round(pct * total / 100.0))
            return '{p:.1f}%  ({v:d})'.format(p=pct, v=val)
        return my_autopct

    fig, ax = plt.subplots()
    ax.pie(sizes, explode=explode, labels=labels, shadow=True,
           autopct=make_autopct(sizes))
    ax.axis('equal')
    
    
df = pd.read_excel("BreastTissue.xls", sheet_name = "Data")
df.drop("Case #", axis = 1, inplace = True)
normalizer = lambda X: preprocessing.MinMaxScaler().fit_transform(X)
df[df.drop("Class",axis =1 ).columns] = normalizer(df.drop("Class",axis =1 ))
```

### Do dataset

```python
df.head()
```

```python
plot_pie(df["Class"])
```

```python
def categorical_toNumerical(y):
    from sklearn import preprocessing
    enconder_output = preprocessing.LabelEncoder()
    a_str = y.astype('str')
    enconder_output.fit(a_str)
    categorical_to_numerical = enconder_output.transform(a_str)
    return(categorical_to_numerical)


df["Class"]  = categorical_toNumerical(df["Class"])

normalizer = lambda X: sklearn.preprocessing.MinMaxScaler().fit_transform(X)


df[df.drop("Class", axis = 1 ).columns] = normalizer(df
                                                   .drop("Class", axis = 1 ))

(X_train, X_test,
 y_train, y_test) = train_test_split(normalizer(df
                                      .drop("Class", axis = 1)), 
                                     df["Class"],
                                     test_size=0.33, shuffle = True)


def nnconvergence(hist, title ="Rede neural - rodada única" ):
    f, ax = plt.subplots(figsize=(14, 10))
    ax.plot(hist.history['categorical_accuracy'], marker = "o", label = "acuracia(teste)")
    ax.plot(hist.history['val_categorical_accuracy'], marker = "o", label = "acuracia(val)")
   # ax.plot(hist.history["loss"], marker = "o", label = "erro de treino")
    ax.set_title(title, fontsize = 20)
    ax.set_ylabel("Medidas", fontsize = 18)
    ax.set_xlabel("Épocas", fontsize = 18)
    ax.legend()
    
```

### Pré-tratamento

- Encoding;
- Normalização;

```python
(X_train, X_test,
 y_train, y_test) = train_test_split(normalizer(df
                                      .drop("Class", axis = 1)), 
                                     df["Class"],
                                     test_size=0.33, shuffle = True)
```

### Implementação em Keras (backend MXNet)

- Dois layers:
 - layer 1 = 32 nodes de output ativação relu
 - layer 2 = 6 nodes de output ativação softmax
- Otimizador = adam
- Função de custo = crossentropy
- Métrica = Acurácia (categórica (one-hot))
- 2500 épocas

```python
model = Sequential()
model.add(Dense(20, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(20, input_dim=X_train.shape[1], activation='tanh'))
model.add(Dense(6, activation='softmax'))
model.compile(optimizer = 'adam',
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])

fittedmodel = model.fit(X_train,
                    keras.utils.to_categorical(y_train),
                    epochs=2500,
                    verbose = False, 
                    validation_data = (X_test, keras.utils.to_categorical(y_test)))
```

```python
titlekeras = ("Rede Neural - rodada única \n Acurácia: $\mu$ = {:.2f} , $\sigma$ = {:.2f}"
                  .format(*list(map(lambda f: f(fittedmodel.history["val_categorical_accuracy"]),
                                    [np.mean, np.std]))))
```

```python
nnconvergence(fittedmodel, titlekeras)
```

## Implementação em Gluon

### Definimos os Iterators (train,test $\rightarrow$ nd.arrays $\rightarrow$ ArrayDataSet $\rightarrow$ DataLoader)

```python
train_data = (gdt.DataLoader(gdt.dataset.ArrayDataset(mx.nd.array(X_train),
                                                      mx.nd.array( y_train)),
                             batch_size =  5))

test_data = (gdt.DataLoader(gdt.dataset.ArrayDataset(mx.nd.array(X_test),
                                                     mx.nd.array(y_test)), 
                            batch_size =  5))
```

### Definimos a rede e o otimizador

```python
mx_net = mx_nn.Sequential()
mx_net.add(mx_nn.Dense(20, activation ='relu'),
           mx_nn.Dense(20, activation ='tanh'),
          mx_nn.Dense(6))

mx_net.initialize()
mx_trainer = gluon.Trainer(mx_net.collect_params(),
                          'adam')
```

### Definimos a função de acurácia

```python
def accuracy(data_iterator, net, ctx=mx.cpu()):
    acc = mx.metric.Accuracy()
    for (data, label) in data_iterator:
        data = data.as_in_context(ctx)
        label = label.as_in_context(ctx)
        output = net(data)
        predictions = mx.nd.argmax(output, axis=1)
        acc.update(preds=predictions, labels=label)
    return acc.get()[1]
```

### Definimos o treinamento da rede

```python
def train(net, trainer, train_data, validation_data, epochs, ctx=mx.cpu()):
    training_accuracies = []
    validation_accuracies = []
    losses = []
    softmax_cross_entropy = gluon.loss.SoftmaxCrossEntropyLoss()
    for e in range(epochs):
        for (data, label) in train_data:
            data  = data.as_in_context(ctx)
            label = label.as_in_context(ctx)
            with autograd.record():
                output = net(data)
                loss = softmax_cross_entropy(output, label)
                loss.backward()
            trainer.step(data.shape[0])
        train_accuracy = accuracy(train_data, net)
        training_accuracies.append(train_accuracy)
        losses.append(loss.asscalar())
        validation_accuracy = accuracy(validation_data, net)
        validation_accuracies.append(validation_accuracy)
       # print("Epoch#{} Time={:.2f} Training={:.4f} Validation={:.4f} Loss={}".format(e,  toc-tic, train_accuracy,
        #                                                                              validation_accuracy,loss.asscalar()))
    return training_accuracies, validation_accuracies, losses
```

### Treinamos a rede

```python
(training_accuracies, 
 validation_accuracies,
 losses) = train(mx_net, mx_trainer,train_data,test_data, 1000)
```

```python
titlegluon = ("Rede Neural - rodada única \n Acurácia: $\mu$ = {:.2f} , $\sigma$ = {:.2f}"
      .format(*list(map(lambda f: f(validation_accuracies), [np.mean, np.std]))))
```

```python
def plot_accuracies(training_accuracies, validation_accuracies, losses, title):
    import matplotlib.ticker as ticker
    epochs = len(training_accuracies)
    plt.clf()
    fig, ax = plt.subplots(figsize=(14, 10))
    plt.xlabel('Épocas', fontsize = 18)
    plt.ylabel('Métricas', fontsize = 18)
    train_plot, = ax.plot(range(epochs), training_accuracies,marker = "o",  label="Training accuracy")
    validation_plot, = ax.plot(range(epochs), validation_accuracies,marker = "o", label="Validation accuracy")
    #loss_plot, = ax.plot(range(epochs), losses,marker = "o", label="Loss")
    plt.title(title, fontsize = 20)
    plt.legend(handles=[train_plot,validation_plot])#, loss_plot])
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
plot_accuracies(training_accuracies, validation_accuracies, losses, titlegluon)
```

# Implementação usando o Module API

### Declaração da rede

```python
net = mx.sym.Variable('data')
net = mx.sym.FullyConnected(net, name='fc1', num_hidden=20)
net = mx.sym.Activation(net, name='relu1', act_type="relu")
net = mx.sym.FullyConnected(net, name='fc2', num_hidden=20)
net = mx.sym.Activation(net, name='tanh1', act_type="tanh")
net = mx.sym.FullyConnected(net, name='fc3', num_hidden=6)
net = mx.sym.SoftmaxOutput(net, name='softmax') #softmax cross entropy loss is added here
```

### Visualização da rede

```python
mx.viz.plot_network(net, node_attrs={"shape":"oval","fixedsize":"false"})
```

### Criamos o Modelo

```python
mod = mx.mod.Module(symbol=net,
                    context=mx.cpu(),
                    data_names=['data'],
                    label_names=['softmax_label'])
```

### Criamos o Iterador

```python
train_iter = mx.io.NDArrayIter(X_train,y_train.values, 5)
val_iter = mx.io.NDArrayIter(X_test, y_test.values, 5)
```

### Treinamos o modelo

```python
import logging
logging.getLogger().setLevel(logging.INFO)


mod.fit(train_iter,
        eval_data=val_iter,
        optimizer='adam',
        eval_metric='acc',
        num_epoch=700 )
```
